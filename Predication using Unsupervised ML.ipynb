## WORKSHOP - 1 K-MEANS CLUSTERING

This workshop moves with some basics of k-means clustering

```python
#importing Essentail Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline

# Loading the IRIS DS

iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
x = iris_df.iloc[:, [0,1,2,3]].values

from sklearn.cluster import KMeans

wcss = []

for i in range (1,11):
    kmeans = KMeans(n_clusters= i, n_init= 10, init= 'k-means++', max_iter=300, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
# plotting in a graph

plt.plot(range(1,11),wcss)
plt.title("Cluster (The elbow method)")
plt.xlabel("no of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters= 3 , random_state=0, n_init=10, init='k-means++', max_iter=300)
y_kmeans = kmeans.fit_predict(x)
```

```python
# Visualising the clusters - On the first two columns
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
